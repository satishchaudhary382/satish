```
principal = 50000
def amort_table(principal, interest, year):
  from prettytable import PrettyTable
  import numpy as np
  rate = (interest/100)/12
  pmt = np.pmt(rate/12, year*12, -principal)
  table = PrettyTable()
  table.field_names = ['Payment No', "Loan_amount", 'Interest', 'PMT', 'Principal','Balance']
  table.add_row(['0',principal,'0', round(pmt,2), '0','0'])
  y = 0
  for i in range(1,(year*12)+1):
    r = principal*(rate/12)
    p = pmt-r
    table.add_row([i,"{:.2f}".format(principal),"{:.2f}".format(r), "{:.2f}".format(pmt),"{:.2f}".format(p), "{:.2f}".format(principal-p)] )
    principal = (principal-p)
    y+=r
  print(table)
  print(f"Total interest Rs {y}")
```

```
amort_table(4500000,7.5, 15)
```
